Ludskym jazykom by sa dal obrazok popisat ako : dva stromy a dom, pripadne jedla, ceresna a dom.

V Jave by to bolo zase takto:

```java
class Jedla {}
class Ceresna {}
class Dom {}
```

Co to ale znamena?
Vezmime prvy riadok:
class Jedla {}

Prve slovo je class co urcuje ze vytvarame triedu, potom nasleduje nazov triedy a potom zoznam vlastnosti, ktore dana trieda ma. Zatial su vsetky bez vlastnosti ale to sa coskoro zmeni.

Co je to ale ta trieda? 
Je to predpis podla ktoreho budeme dalej vytvarat objekty.

Dalme si to radsej a priklade:

Jedla, ked nic viac k tomu nepoviem, je vseobecne hocijaka jedla, ktorych moze byt vela. <br>Je to trieda.

Ale ked definujem, ze je to ta jedla vedla polnej cesty, 2 rocna, ta co je zo severu lysa, tak je to jedna konkretna jedla.<br> Je to objekt.<br>Je to instancia triedy Jedla. 

Ale este sme neskoncili, ked budeme pokracovat v povodnom plane opisat obrazok, budeme potrebovat este dalsie prikazy.<br>
Pretoze zatial sme vytvorili len ten vseobecny predpis a teraz musime este vytvorit z tych predpisov(tried) objekty.

```java
Jedla jedla = new Jedla();
```

Tento prikaz robi to, ze vytvori objekt typu Jedla, to urobi prikaz 'new Jedla()'.<br>
Tento novy objekt potom priradi do premennej ktora sa vola 'jedla' a je ties typu Jedla.<br>
Na dve vety je to trochu vela jedli, ale zatial davam do pozornosti ze Java je case-sensitive, co znamena ze zalezi na velkosti pismen, teda 'jedla', 'Jedla' a 'JEDLA' su tri rozdielne veci.

<br>Ale co je to premenna?
<br>Je to meno, ktore priradime nejakemu objektu aby sme s nim mohli dalej pracovat.
<br>Na priklade:


```java
class Strom {}
Strom jelsa = new Strom();
Strom dub = new Strom();
```

Na ilustraciu som vytvoril novu triedu Strom. <br> 
A dalej som vytvoril dva objekty typu Strom. <br>
Prvy som ulozil do premennej jelsa a druhy do premennej dub.

Takymto sposobom vieme vytvorit svoju vlastnu triedu, podla toho aky problem chceme namodelovat. <br>
Existuje ale vela preddefinovanych tried, ktore uz su sucastou jazyka Java, ktore tvorcovia ocakavali ze budu uzitocne.
Napr trieda java.lang.String, ktora sa pouziva na ulozenie znakov, a my ju aj coskoro pouzijeme.

Ale aby sme si uzavreli povodne zadanie, opisat obrazok na zaciatku notebooku.<br>Kod moze vyzerat takto


```java
// toto je komentar, pouziva sa na doplnkove informacie ku kodu, nieje sucastou kodu 
// teda najprv si zadefinujeme triedy z ktorych potom vytvorime objekty dom a strom
class Strom {}
class Dom {}
```
